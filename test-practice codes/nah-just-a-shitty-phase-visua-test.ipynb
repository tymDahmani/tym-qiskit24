```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector

# Create a quantum circuit
circuit = QuantumCircuit(1)

# Apply the Z gate
circuit.z(0)

# Simulate the circuit with statevector simulator before applying Z gate
simulator = Aer.get_backend('statevector_simulator')
result_before = execute(circuit, simulator).result()
statevector_before = result_before.get_statevector()

# Apply the Z gate
circuit.z(0)

# Simulate the circuit with statevector simulator after applying Z gate
result_after = execute(circuit, simulator).result()
statevector_after = result_after.get_statevector()

# Visualize the Bloch vectors before and after
plot_bloch_multivector(statevector_before, title='Before Z gate')
```

```python

plot_bloch_multivector(statevector_after, title='After Z gate')
```
