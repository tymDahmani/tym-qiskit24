```python
wtamiworth = QuantumCircuit(6,4)
```

```python
def phase_kickback(q):
    
    # q.x(3)
    # q.ccx(2,3,5)
    # q.ccx(1,5,0)
    # q.cx(0,4)
    # q.ccx(1,5,0)
    # q.ccx(2,3,5)
    # q.x(3)
    
#     phase kickback using z gates for the winner state |0101>
    # prepare the winner qubits for the z gates
    q.x([0,2])
    
#     apply the phase kickback
    q.cz(0,1)
    q.cz(2,3)
    
    
    q.barrier()

# def phase_kickback(q):
#     q.x([0, 2])  # Flip the bits corresponding to the target state
#     q.cz(0, 2)      # Apply a controlled-Z gate between the first and third qubits
#     q.x([0, 2])  # Undo the bit flips
#     q.barrier()

    
def ancilla_prep(q):
    q.x(4)
    q.h(4)
    
    q.barrier()

def inversion(q):
    q.h([0,1,2,3])
    q.x([0,1,2,3])
    
    q.ccx(0,1,2)
    q.ccx(2,3,4)
    
    q.barrier()
    q.x([0,1,2,3])
    q.h([0,1,2,3])
    
    q.barrier()
    
#     for ancilla
    q.x(4)
    
#     set ancilla back
    q.h(4)
    q.x(4)
    
    q.barrier()
    
def iterations(circ):
    phase_kickback(circ)
    inversion(circ)
    
```

```python
wtamiworth.h([0,1,2,3])

wtamiworth.barrier()

ancilla_prep(wtamiworth)

# 1st iteration
iterations(wtamiworth)

# 2nd iteration
iterations(wtamiworth)

# 3rd iteration
iterations(wtamiworth)

# 4th iteration
iterations(wtamiworth)

# 5th iteration
iterations(wtamiworth)

# 6th iteration
iterations(wtamiworth)

# 7th iteration
iterations(wtamiworth)

# 8th iteration
iterations(wtamiworth)

wtamiworth.barrier()

wtamiworth.measure([0,1,2,3],[0,1,2,3])
```
