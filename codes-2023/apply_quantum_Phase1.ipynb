```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_bloch_multivector

# Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)

# Apply a phase shift of pi/4 (45 degrees) using the Rz gate
qc.y(0)  # 0.25 * pi

# Visualize the circuit
print(qc)

# Simulate the circuit on a quantum simulator
simulator = Aer.get_backend('statevector_simulator')
compiled_circuit = transpile(qc, simulator)
job = assemble(compiled_circuit, shots=1)
result = simulator.run(job).result()
statevector = result.get_statevector()

# Print the final statevector
print("Final statevector:", statevector)

# Visualize the state on the Bloch sphere
plot_bloch_multivector(statevector)
```
