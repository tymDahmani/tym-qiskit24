```python
from qiskit import *
```

```python
cr = QuantumCircuit(2, 2)
```

```python
cr.draw()
```

```python
cr.x([0, 1])
cr.h(1)
```

```python
cr.draw()
```

```python
cr.measure([0, 1], [0, 1])
```

```python
cr.draw()
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(cr, backend = simulator).result()
```

```python
# from qiskit.tools.visualization import plot_histogram
# plot_histogram(result.get_counts(cr))
from qiskit.visualization import plot_bloch_multivector

statevector = result.get_statevector()
plot_bloch_multivector(statevector)
```

```python

# Create a quantum circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
#### never measure the circuit if 

# Choose the UnitarySimulator backend
simulator = Aer.get_backend('unitary_simulator')

# Execute the circuit and retrieve the unitary matrix
result = execute(qc, backend=simulator).result()
unitary_matrix = result.get_unitary()

# Print the unitary matrix
print(unitary_matrix)
```
